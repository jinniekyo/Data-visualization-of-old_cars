```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv('old_cars.csv')

# Hitung rata-rata MPG untuk setiap wilayah asal
mean_mpg = df.groupby('Origin')['MPG'].mean()

# Bagi nilai rata-rata MPG menjadi beberapa kategori dengan kenaikan 2 mpg
bins = range(10, 50, 2)
labels = [f'{i}-{i+2}' for i in bins[:-1]]
df['mpg_group'] = pd.cut(df['MPG'], bins=bins, labels=labels)

# Hitung jumlah model yang termasuk ke dalam setiap kategori MPG dan wilayah asal
grouped = df.groupby(['Origin', 'mpg_group']).size().unstack()
```

```python
df.info()
```

```python
# Buat grouped bar chart
ax = grouped.plot(kind='bar', width=0.8, figsize=(12,8))
ax.set_xlabel('Number of Cars')
ax.set_ylabel('MPG Range')
ax.set_title('Distribution of MPG by Origin')
ax.legend(['US', 'Europe', 'Japan'])
plt.show()
```

```python
df = df[['Origin', 'Model', 'MPG']]
```

```python
df = df.groupby(['Origin', 'Model'])['MPG'].mean().reset_index()
```

```python
import matplotlib.pyplot as plt

# Buat plot untuk setiap asal geografis
for origin in df['Origin'].unique():
    # Filter data hanya untuk asal geografis tertentu
    data = df[df['Origin'] == origin]

    # Buat line chart
    plt.plot(data['Model'], data['MPG'], label=origin, )

# Berikan label pada sumbu x, y, dan judul
plt.xlabel('Model')
plt.ylabel('MPG')
plt.title('Evolusi Temporal MPG untuk Setiap Asal Geografis')

# Tampilkan legenda
plt.legend()

# Tampilkan plot
plt.show()
```

```python
df = pd.read_csv('old_cars.csv')

# Create scatter plot
plt.scatter(df['Horsepower'], df['MPG'], c=df['Model'], cmap='viridis')

# Add labels and title
plt.xlabel('Horsepower')
plt.ylabel('MPG')
plt.title('Relationship between Horsepower and MPG with Year Code Color')

# Add colorbar
cbar = plt.colorbar()
cbar.set_label('Model')

# Show the plot
plt.show()
```

```python
# membaca file csv
df = pd.read_csv('old_cars.csv')


# load dataset 'mpg' dari seaborn
df = sns.load_dataset('mpg')

# membuat objek PairGrid
pg = sns.PairGrid(df, vars=['mpg', 'weight', 'horsepower', 'displacement'], hue='origin', palette='Set1')

# menambahkan scatter plot ke setiap subplot pada PairGrid
pg.map(sns.scatterplot)

# menambahkan legenda
pg.add_legend()

# menampilkan plot
plt.show()
```
